```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

input_file = "../out/output.csv"
data = pd.read_csv(input_file)
data = data.sort_values("Vertices")
x = np.sort(data["Vertices"].to_numpy())
seq = data["SequentialTime"].to_numpy()
openMP = data["OpenMpTime"].to_numpy()
Cuda = data["CudaTime"].to_numpy()
print(x)

plt.plot(x, seq, color="red", label="Sekwencyjnie")
plt.plot(x, openMP, label="OpenMP")
plt.plot(x, Cuda, color="green", label="CUDA")

plt.title("Czas obliczeń dla różnych algorytmów")
plt.xlabel("Vertices")
plt.ylabel("Czas obliczeń [ms]")

plt.legend()
# plt.show()
plt.savefig("wykres.png")



display(data.sort_values("Vertices"))
```
